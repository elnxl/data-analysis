Мешалкин Сборник задач по теории вероятностей n 59. 
В квадрат с вершинами (0, 0), (0, 1), (1,0), (1,1) наудачу брошена точка М. Пусть (x, y) будут ее координаты. Предполагается, что вероятность попадания в область, лежащую целиком внутри квадрата, зависит лишь от площади этой области и пропорциональна ей.
Найти для z = 1/4

Вариант 3
P{min(x,y)<z}

Импортируем библиотеки

```
import random
```

В цикле случайно получаем числа [0.0, 1.0). 
В условии if проверяем условие задачи и, если оно выполняется, увеличиваем result.
Вероятность данного события, разделив результат на число экспериментов

```
number_experiments = 100000
z = 1/4

random.seed(a=42, version=2)
result = 0 
for el in range(number_experiments):
    x = random.random()
    y = random.random()
    if  min(x, y)< z:
        result += 1
    probability = result/number_experiments
print('Win probability = {:.4f}'.format(probability))
```

Проверим результат, увеличив число экспериментов в 2 раза

```
number_experiments = 200000
z = 1/4

random.seed(a=42, version=2)
result = 0 
for el in range(number_experiments):
    x = random.random()
    y = random.random()
    if  min(x, y)< z:
        result += 1
    probability = result/number_experiments
print('Win probability = {:.4f}'.format(probability))
```

Вывод: вероятность, посчитанная с разным числом экспериментов, прирост в их количеств дает более точные результаты, но в процентном соотношении это всего лишь 0.1%
