Колода игральных карт содержит 52 карты, разделяющиеся на 4 различные масти по 13 карт в каждой. Предположим, что колода тщательно перетасована, так что вытаскивание любой карты одинаково вероятно. Вытащим 6 из них. 

Вариант 3
Найти вероятность того, что среди этих карт королей будет больше чем десяток

Импортируем библиотеки.

```
import random
```

deck - массив для карт, с картами: король, 10 и остальне, нам не важные
При выборе 6 карт, мы проверям если число королей больше числа 10, но в result +1, далее считаем вероятность данного события, разделив результат на число экспериментов

```
random.seed(a=42, version=2)
number_experiments = 100000
deck = ['king'] * 4 + ['10'] * 4 + ['s'] * 44 
result = 0
for i in range(number_experiments):
    a = random.sample(deck, 6)                      # Выбираем 6 карт без возвращения
    if a.count('king') > a.count('10'):
      result += 1
probability = result/number_experiments             # Оцениваем вероятность

print('Win probability = {:.4f}'.format(probability))
```

Проверим результат, увеличив число экспериментов в 2 раза

```
random.seed(a=42, version=2)
number_experiments = 200000
deck = ['king'] * 4 + ['10'] * 4 + ['s'] * 44 
result = 0
for i in range(number_experiments):
    a = random.sample(deck, 6)                      # Выбираем 6 карт без возвращения
    if a.count('king') > a.count('10'):
      result += 1
probability = result/number_experiments             # Оцениваем вероятность

print('Win probability = {:.4f}'.format(probability))
```

Сравним результат с помощью встроенной функции для числа сочетаний

```
from scipy.special import comb
```

```
number = comb(52,6)

result = 0
for i in range(6):
  for j in range(i+1,7-i):
    result += comb(4,i)*comb(4,j)*comb(52-8,6-i-j)
probability = result/number
print('Win probability = {:.4f}'.format(probability))
```

Вывод: вероятность посчитанная 2 способами и с разным числом экспериментов дала достаточно точно похожие результаты, с разность в 0.01%
