Задание: В лотерее 100 билетов, из которых 5 выигрышные. Как велика вероятность выигрыша для того, кто имеет 15 билетов?

Импортируем библиотеки.

```
import numpy as np
import random
import matplotlib.pyplot as plt
```

Входные данные

```
all_tickets = 100
win_tickets = 5
choоsen_tickets = 15
```

win －лист случайно выбранных выигрышных билетов
choosen - лист случайно выбранных билетов
Сравнение двух листов на наличие выигрышных. Если имеется совпадение, значит вытянут выигрышный билет. При этом check становится равным true. В конце цикла смотрится значение check и если оно true - к  result (количество выигрышей) добавляется 1. 
Для подсчета вероятности количество выигрышей делится на общеe число экспериментов.

```
number_experiments = 100000
random.seed(a=42, version=2)
A = list(range(1,101))
result = 0
prob_tmp = []

for el in range(number_experiments):
  check = False
  win = random.sample(A, k = win_tickets)
  choosen = random.sample(A, k = choоsen_tickets)
  for i in range(0, len(win)):
    for j in range(0, len(choosen)):
      if win[i] == choosen[j]:
        check = True
  if check:
    result += 1
  prob_tmp.append(result/(el+1))
print('Win probability = {:.4f}'.format(result/number_experiments))
plt.plot(prob_tmp[10000 : 100000])
```

Проверим результат, увеличив число экспериментов в 2 раза

```
number_experiments = 200000
random.seed(a=42, version=2)
A = list(range(1,101))
result = 0
prob_tmp = []

for el in range(number_experiments):
  check = False
  win = random.sample(A, k = win_tickets)
  choosen = random.sample(A, k = choоsen_tickets)
  for i in range(0, len(win)):
    for j in range(0, len(choosen)):
      if win[i] == choosen[j]:
        check = True
  if check:
    result += 1
  prob_tmp.append(result/(el+1))
print('Win probability = {:.4f}'.format(result/number_experiments))
plt.plot(prob_tmp[10000 : 200000])
```

Сравним результат с помощью встроенной функции для числа сочетаний

```
from scipy.special import comb
```

```
result = 0
for el in range(1, choоsen_tickets+1):
  result += comb(win_tickets,el)*comb(all_tickets-win_tickets,choоsen_tickets-el)/comb(all_tickets,choоsen_tickets)
print('Win probability = {:.4f}'.format(result))
```

Вывод: вероятность приблизилась к значению, полученному с помощью встроенной функции. Соответственно, увеличение числа экспериментов дало лучшую точность результата, но в процентном соотношении это всего лишь 0.1%
